```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df_sports=pd.read_excel('/content/drive/MyDrive/data_sports.xlsx')
df_sports
```

```python
fig = plt.figure(dpi=100)
plt.plot(df_sports.iloc[0,1:], 'r*--', label=df_sports.iloc[0,0])
plt.plot(df_sports.iloc[1,1:], 'bs-', label=df_sports.iloc[1,0])
plt.plot(df_sports.iloc[2,1:], 'g-D', label=df_sports.iloc[2,0])
plt.title("What is America's most popular sport?")
plt.xlabel("Years")
plt.ylabel("Percentage of spectators")
plt.legend(loc='upper right')
plt.axis([2003,2018,0,55])

plt.show()
```
